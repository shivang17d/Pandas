```python
def max_runs_scored_against_team(batsman_name):
    vk = df[df['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
max_runs_scored_against_team('V Kohli')
```

```python
max_runs_scored_against_team('MS Dhoni')
```
